### 시계열 분석

------

$ACF(\text{lag})$=$\frac{\sum(y_t - \bar{y})(y_{t-\text{lag}} - \bar{y})}{\sum(y_t - \bar{y})^2}$

여기서,
- $ACF(\text{lag})$는 시간 지연(\text{lag})에 대한 자기상관 계수를 나타냅니다.
- $y_t$는 시계열 데이터의 현재 시간 $t$에서의 관측값을 나타냅니다.
- $\bar{y}$는 시계열 데이터의 평균값을 나타냅니다.
- $y_{t-\text{lag}}$는 현재 시간 $t$에서 $\text{lag}$ 시간 단계 이전의 관측값을 나타냅니다.

$PACF(lag)$ = $Corr(y_t, y_{t-lag} | y_{t-1}, y_{t-2}, ..., y_{t-lag+1})$

여기서,
- $PACF(lag)$는 시간 지연($lag$)에 대한 부분 자기상관 계수를 나타냅니다.
- $y_t$는 현재 시간 $t$에서의 관측값을 나타냅니다.
- $y_{t-lag}$는 현재 시간 $t$에서 $lag$ 시간 단계 이전의 관측값을 나타냅니다.
- $y_{t-1}$, $y_{t-2}$, ..., $y_{t-lag+1}$는 현재 시간 이전의 관측값들을 나타냅니다.


---------

AR(n)은 자기회귀(AutoRegressive) 모델의 표기법 중 하나로, n 시간 단계 이전의 자기상관된 데이터 포인트를 사용하여 현재 시점의 데이터를 예측하는 시계열 모델을 나타냅니다. AR(n) 모델의 일반적인 수식은 다음과 같습니다:

\[ X_t = c + \phi_1 X_{t-1} + \phi_2 X_{t-2} + \ldots + \phi_n X_{t-n} + \varepsilon_t \]

여기서,
- \( X_t \)는 현재 시간 \( t \)에서의 시계열 데이터 포인트를 나타냅니다.
- \( c \)는 상수항을 나타냅니다.
- \( \phi_1, \phi_2, \ldots, \phi_n \)은 AR 모델의 파라미터로, 각각 n 시간 단계 이전의 자기상관된 데이터 포인트 \( X_{t-1}, X_{t-2}, \ldots, X_{t-n} \)에 대한 가중치를 나타냅니다.
- \( \varepsilon_t \)는 현재 시간 \( t \)에서의 백색 잡음(white noise) 오차 항을 나타냅니다.

MA(n)은 이동평균(Moving Average) 모델의 표기법 중 하나로, n 시간 단계 전의 오차 항을 사용하여 현재 시점의 데이터를 예측하는 시계열 모델을 나타냅니다. MA(n) 모델의 일반적인 수식은 다음과 같습니다:

\[ X_t = c + \varepsilon_t + \theta_1\varepsilon_{t-1} + \theta_2\varepsilon_{t-2} + \ldots + \theta_n\varepsilon_{t-n} \]

여기서,
- \( X_t \)는 현재 시간 \( t \)에서의 시계열 데이터 포인트를 나타냅니다.
- \( c \)는 상수항 또는 평균을 나타냅니다.
- \( \varepsilon_t \)는 현재 시간 \( t \)에서의 백색 잡음(white noise) 오차 항을 나타냅니다.
- \( \theta_1, \theta_2, \ldots, \theta_n \)은 MA 모델의 파라미터로, 각각 n 시간 단계 전의 오차 항 \( \varepsilon_{t-1}, \varepsilon_{t-2}, \ldots, \varepsilon_{t-n} \)에 대한 가중치를 나타냅니다.
