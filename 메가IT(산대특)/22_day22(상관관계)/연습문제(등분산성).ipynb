```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy as cp
```

```python
plt.rcParams["font.size"]=10
plt.rcParams["figure.figsize"]=(10,6)
plt.rcParams["font.family"]='BM YEONSUNG'
plt.rcParams["axes.unicode_minus"]=False
```

#### 틀

## 문제1

```python
Data=pd.read_excel('https://data.hossam.kr/E03/baseball.xlsx')
data=pd.DataFrame(Data)
```

```python
data
```

```python
cols=[data.columns]
```

```python
plt.rcParams["figure.figsize"]=(10,5)
f,a=plt.subplots(len(cols),2)
for i in range(len(cols)):
    sb.heatmap(data=data.filter(cols[i]).corr(method="pearson"),
                       cmap="Greens",
                       vmin=-1,
                       vmax=1,
                       annot=True,
                       ax=a[0])
    sb.regplot(data=data.filter(cols[i]),
                       x=cols[i][0],
                       y=cols[i][1],
                       ax=a[1])
plt.show()
plt.close()
```
