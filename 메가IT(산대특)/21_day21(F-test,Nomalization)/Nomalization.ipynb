```python
import  pandas as pd
from sklearn import preprocessing as pp
```

```python
Data=pd.read_excel('https://data.hossam.kr/D05/gradeuate.xlsx')
data=pd.DataFrame(Data)
```

```python
data.head(2)
```

## Min-Max-scaler

```python
min=data['학부성적'].min()
max=data['학부성적'].max()
```

```python
data["min-max(1)"]=data["학부성적"].apply(lambda x:(x-min)/(max-min))
```

```python
scale=pp.MinMaxScaler()
scale.fit(data[["학부성적"]])
data['min-max(2)']=scale.transform(data[["학부성적"]])
```

```python
data
```

## Z-score(standardization)

```python
std=data["학부성적"].std()
average=data['학부성적'].mean()
```

```python
data["Z-score(1)"]=data["학부성적"].apply(lambda x:(x-average)/std)
```

```python
scale2=pp.StandardScaler()
scale2.fit(data[["학부성적"]])
data["Z-score(2)"]=scale2.transform(data[["학부성적"]])
```

```python
data
```

## Robust
> 사분위수를 이용

```python
Q1=data["학부성적"].quantile(0.25)
Q3=data["학부성적"].quantile(0.75)
Q2=data["학부성적"].quantile(0.5)
```

```python
data["Robust(1)"]=data["학부성적"].apply(lambda x:(x-Q2)/(Q3-Q1))
```

```python
scale3=pp.RobustScaler()
scale3.fit(data[["학부성적"]])
data["Robust(2)"]=scale3.transform(data[["학부성적"]])
```

```python
data
```
