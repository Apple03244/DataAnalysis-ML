```python
info['query']="봄 패션"
info['size']=80
```

```python
isend=False
info['page']=1
count=0
```

```python
while not isend:
    url=f"https://dapi.kakao.com/v2/search/image?query={info['query']}&page={info['page']}&size={info['size']}"
    result=session.get(url)
    if result.status_code!=200:
        raise Exception(f'{result.status_code}{result.reason}')
    result.encoding='utf-8'
    data=json.loads(result.text)

    isend=data['meta']['is_end']
    info['page']+=1

    if not os.path.exists(info['query']):
        os.mkdir(info['query'])
    for i,v in enumerate(data['documents']):
        count+=1
        session.headers.update({
            'Authorization' : "KakaoAK 4fd65a1fc1d9de84fb9809d081f3fc8c",
            'Refer' : v['doc_url'],
            'User_Agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
        })
        try:
            result=session.get(v["image_url"],stream=True)
        except:
            continue
        if result.status_code!=200:
            continue
        file_name="%s/%05d.png" % (info['query'],count)
        with open(file_name,'wb') as f:
            f.write(result.raw.read())
            print("Save_successful")
print('okay')
```
