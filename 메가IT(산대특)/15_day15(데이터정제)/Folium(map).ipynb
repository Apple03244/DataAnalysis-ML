```python
import folium
import pandas as pd
```

```python
url='https://www.google.co.kr/maps/place/%EA%B2%BD%EA%B8%B0%EB%8F%84+%EA%B9%80%ED%8F%AC%EC%8B%9C+%ED%83%9C%EC%9E%A5%EB%A1%9C+845/data=!3m1!4b1!4m6!3m5!1s0x357c86b69fdcf9bb:0x8b59d720c253d840!8m2!3d37.6334672!4d126.6780971!16s%2Fg%2F11sb78841r?hl=ko&entry=ttu'
```

```python
url.find("!3d")
```

### 위도 경도 추출

```python
a,b=url[url.find("!3d")+3:url.find("!4d")],url[url.find("!4d")+3:url.find("!16s")] #구글 맵 url에서 싸옴
```

### 맵 구현

```python
map_osm=folium.Map(location=[a,b],zoom_start=13)
```

### 마커 찍기

```python
maker=folium.Marker([a,b],popup="스위트홈",
                    icon=folium.Icon(color="dark"))
maker.add_to(map_osm)
```

```python
map_osm
```

### 사용자 아이콘 넣기

```python
pp=folium.Popup("<dir style='white-space :nowrap'><h3>스위트홈</h3></dir>")
icn_img=folium.features.CustomIcon('C:/Users/SAMSUNG/Desktop/애용.png',icon_size=(150,150))
user_maker=folium.Marker([a,b],popup=pp,icon=icn_img)
user_maker.add_to(map_osm)
```

## 서울시 고등학교 모두 보기(실습)

```python
D=pd.read_excel("https://data.hossam.kr/D04/school.xlsx")
data=pd.DataFrame(D)
```

```python
pre_data=data.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
```

```python
final_data=pre_data.query('학교급구분=="고등학교" and 소재지도로명주소.str.contains("서울")')
```

```python
map_osm=folium.Map(location=[37.566651, 126.978428], zoom_start=14)

for i in final_data.index:
    maker=folium.Marker([final_data.loc[i]["위도"],final_data.loc[i]["경도"]],
                        popup=f"<font color='Blue' style='white-space: nowrap'><b>{final_data.loc[i]['학교명']}</b></font>")
    maker.add_to(map_osm)
```
