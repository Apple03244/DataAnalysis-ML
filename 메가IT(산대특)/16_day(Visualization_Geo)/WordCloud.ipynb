#### 금지어 설정

```python
A=STOPWORDS
A.add("sail")
A.add("Alice")
```

```python
img=Img.open("./앨리스배경.png")
arry_img=np.array(img)
```

```python
X=[]
for x in arry_img:
    X.append(list(map(lambda x: 0 if x==255 else 255,x)))
```

```python
test_X=np.array(X)
```

```python
wc=WordCloud(width=1000,height=800,max_font_size=150,scale=2.0,max_words=120,stopwords=A,mask=test_X)
wc.generate(T)
wc.to_file("test.png")
display(Image("./test.png"))
```

```python
def getColor(word,font_size,position,orientation,random_state=None,**kwargs):
    color="yellow"
    random_state.seed()                     # 랜덤숫자 카드를 섞는다.
    a = random_state.randint(30, 90)        # 채도를 랜덤하게 설정
    b = random_state.randint(30, 90)        # 명도를 랜덤하게 설정
    color = "hsl(270, %d%%, %d%%)" % (a, b)
    return color
```

```python
wc=WordCloud(width=1000,height=800,max_font_size=150,scale=2.0,max_words=120,stopwords=A,mask=test_X,color_func=getColor)
wc.generate(T)
wc.to_file("test.png")
display(Image("./test.png"))
```
