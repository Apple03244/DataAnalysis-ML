```python
plt.figure()
plt.title("문제 5_1 (2015년)")
plt.barh(x+0.15,data4[2015],label="2015",height=0.3)
plt.barh(x-0.15,data4[2016],label="2016",height=0.3)
plt.legend()
plt.yticks(x,x_label)
plt.show()
plt.close()
```

```python
D=pd.read_excel("https://data.hossam.kr/D01/traffic_acc.xlsx")
data5=pd.DataFrame(D)
```

```python
data5.dtypes
```

```python
X=data5["발생건수"]
Y=data5["부상자수"]
```

```python
plt.figure()
plt.title("발생건수와 부상자 수의 상관관계")
plt.scatter(X,Y)
plt.show()
plt.xlabel("부상자수")
plt.ylabel("발생건수")
plt.close()
```
