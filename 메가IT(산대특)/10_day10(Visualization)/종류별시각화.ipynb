#### Bar chart

가로막대

```python
x=["Seoul","Busan","Jeju"]
y=[20,30,10]
```

```python
plt.figure()
plt.title("Bar chart 가로")
plt.bar(x,y,color="Blue")
plt.xlabel("지역")
plt.ylabel("지역별 매출")
plt.show()
plt.close()
```

세로 막대 그래프

```python
plt.figure()
plt.title("Bar chart 가로")
plt.barh(x,y,color="Blue")
plt.xlabel("지역")
plt.ylabel("지역별 매출")
plt.show()
plt.close()
```

#### 다중 막대 그래프

```python
data=pd.read_excel("https://data.hossam.kr/grade.xlsx")
```

```python
data
```

```python
# x축
x_label=list(data["이름"])
```

```python
# numpy로 처리해보자.
X=np.array(range(len(x_label)))
```

```python
#y 축
Y1=list(data["국어"])
Y2=list(data["영어"])
```

```python
plt.figure()
plt.title("Multi Bar chart")
plt.bar(X-0.15,Y1,color="Blue",label="국어",width=0.3)
plt.bar(X+0.15,Y2,color="Red",label="영어",width=0.3)
plt.xlabel("학생")
plt.ylabel("점수")
plt.xticks(X,x_label)
plt.legend()
plt.grid(alpha=1,axis="y",which="major")
plt.grid(alpha=0.5,axis="y",which="minor")
plt.show()
plt.close()
```

#### Pie chart

```python
data=[10,20,30,70]
labels=["봄",'여름','가을',"겨울"]
colors=['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode=(0,0.1,0,0)
```

```python
plt.figure()
plt.title("Pie chart")
plt.pie(
   data,colors=colors,labels=labels,
   autopct="%0.1f%%",explode=explode,
   startangle=60
)
plt.show()
plt.close()
```
