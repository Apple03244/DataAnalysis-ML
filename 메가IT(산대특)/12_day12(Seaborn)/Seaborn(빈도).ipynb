```python
import matplotlib.pyplot as plt
import statsmodels.graphics.mosaicplot as msc
import pandas as pd
import seaborn as sb
import numpy as np
plt.rcParams["font.family"] = 'BM YEONSUNG'
#plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (5, 5)
plt.rcParams["axes.unicode_minus"] = False
```

```python
D=pd.read_excel("https://data.hossam.kr/D01/titanic.xlsx")
data=pd.DataFrame(D)
# data.drop("PassengerId",inplace=True)
```

```python
data.drop(columns=["PassengerId"],inplace=True)
```

```python
data=data.astype({'Survived':'category',
             'Pclass':'category',
             'Cabin':'category',
             'Embarked':'category'})
```

catplot의 그래프 사이즈는 객체를 리턴받아 개별적으로 맞춰주어야한다.

```python
g=sb.catplot(
    data=data,
    x="Pclass",
    kind="count",
    palette="pastel",
    edgecolor="0.3"
)
g.fig.set_figwidth(5)
g.fig.set_figheight(3)
plt.show()
```

범례추가

```python
g=sb.catplot(
    data=data,
    x="Pclass",
    hue="Sex",
    kind="count",
    palette="pastel",
    edgecolor="0.3"
)
g.fig.set_figwidth(5)
g.fig.set_figheight(3)
plt.show()
```

```python
g=sb.catplot(
    data=data,
    x="Sex",
    hue="Pclass",
    kind="count",
    palette="pastel",
    edgecolor="0.3"
)
g.fig.set_figwidth(5)
g.fig.set_figheight(3)
plt.show()
```

아예 그래프를 분리하고 싶으면
병렬 빈도 막대 그래프를 이용해라

```python
g=sb.catplot(
    data=data,
    x="Pclass",
    hue="Sex",
    kind="count",
    palette="pastel",
    edgecolor="0.3",
    col="Survived"
)
g.fig.set_figwidth(5)
g.fig.set_figheight(3)
plt.show()
```

가로 빈도 막대 그래프

```python
g=sb.catplot(
    data=data,
    y="Pclass",
    hue="Sex",
    kind="count",
    palette="pastel",
    edgecolor="0.3",
    col="Survived"
)
g.fig.set_figwidth(5)
g.fig.set_figheight(3)
plt.show()
```
