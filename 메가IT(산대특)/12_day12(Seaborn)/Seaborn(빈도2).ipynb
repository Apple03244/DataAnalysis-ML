```python
import matplotlib.pyplot as plt
import statsmodels.graphics.mosaicplot as msc
import pandas as pd
import seaborn as sb
import numpy as np
plt.rcParams["font.family"] = 'BM YEONSUNG'
#plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["axes.unicode_minus"] = False
```

```python
titanic = pd.read_excel("https://data.hossam.kr/D01/titanic.xlsx", index_col="PassengerId")
df = titanic.astype({"Survived": "category", "Pclass": "category", "Sex": "category", 
                     "Cabin": "category", "Embarked": "category"})
```

```python
sb.countplot(data=df,x="Sex",hue="Pclass")
plt.xticks([0,1],["남자","여자"]) #무조건 0부터임
plt.show()
```

mosaic plot

```python
def getColor(key):
    if "1" in key:
        color="Red"
    if "2" in key:
        color="Yellow"
    if "3" in key:
        color="Green"
    return {"color":color}
```

```python
def getLabel(x):
    if "female" in x:
        name="여자"
    else:
        name="남자"
    if "1" in x:
        clas="1등석"
    elif "2" in x:
        clas="2등석"
    else:
        clas="1등석"
    return f'{name} & {clas}'
    
```

```python
msc.mosaic(df, ['Pclass',"Sex"], axes_label=False,
           properties=getColor,
           labelizer=getLabel)

plt.title('타이타닉호 클래스 비율', fontsize=17)
plt.show()
plt.close()
```
