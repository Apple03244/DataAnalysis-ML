```python
import matplotlib.pyplot as plt
import statsmodels.graphics.mosaicplot as msc
import pandas as pd
import seaborn as sb
import numpy as np
plt.rcParams["font.family"] = 'BM YEONSUNG'
#plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["axes.unicode_minus"] = False
```

```python
Data=pd.read_excel('https://data.hossam.kr/D01/mpg.xlsx')
data=pd.DataFrame(Data)
```

문제0. 전처리

```python
data.head(2)
```

```python
data.drop(columns=["id"],inplace=True)
```

```python
data.dtypes
```

문제 1

```python
sb.countplot(data=data,
             x=data["manufacturer"],
             palette='YlGnBu')
plt.xticks(rotation=45)
plt.show()
```

문제2

```python
sb.countplot(data=data,
             x=data["manufacturer"],
             palette='YlGnBu',
             hue="year")
plt.xticks(rotation=45)
plt.legend(loc=9)
plt.show()
```

```python
sb.countplot(data=data,
             x=data["drv"],
             palette='YlGnBu',
             hue="year")
plt.xticks([0,1,2],["전륜구동","사륜구동","후륜구동"])
plt.legend(loc=1)
plt.show()
```

문제4.

```python
sb.barplot(
    data=data,
    x=data["drv"],
    y=data['cty'],
    estimator='average',
    palette='YlGnBu',
    ci=None
)
plt.xticks([0,1,2],["전륜구동","사륜구동","후륜구동"])
plt.show()
```

문제5.

```python
plt.rcParams["figure.figsize"] = (12, 6)
fig,(a1,a2)=plt.subplots(1,2)

sb.barplot(ax=a1,
    data=data,
    x=data["manufacturer"],
    y=data['cty'],
    palette='YlGnBu',
    ci=None
)
# plt.xticks(rotation=90,ax=a1)
sb.barplot(ax=a2,
    data=data,
    x=data["manufacturer"],
    y=data['hwy'],
    estimator='average',
    palette='YlGnBu',
    ci=None
)
plt.xticks(rotation=90)
plt.show()
```

문제6.

```python
plt.rcParams["figure.figsize"] = (12, 6)
fig,(a1,a2)=plt.subplots(1,2)

sb.scatterplot(ax=a1,
    data=data,
    x=data["displ"],
    y=data['cty'],
    palette='YlGnBu'
)
sb.regplot(ax=a1,
    data=data,
    x=data["displ"],
    y=data['cty'],
)
# plt.xticks(rotation=90,ax=a1)
sb.scatterplot(ax=a2,
    data=data,
    x=data["displ"],
    y=data['hwy'],
    palette='YlGnBu'
)

sb.regplot(ax=a2,
    data=data,
    x=data["displ"],
    y=data['hwy']
)
plt.show()
```
