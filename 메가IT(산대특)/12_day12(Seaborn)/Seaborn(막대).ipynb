```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
plt.rcParams["font.family"] = 'BM YEONSUNG'
#plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (7, 4)
plt.rcParams["axes.unicode_minus"] = False
```

```python
D=pd.read_excel('https://data.hossam.kr/D01/grade.xlsx')
data=pd.DataFrame(D)
```

```python
plt.rcParams["font.family"] = 'BM YEONSUNG'
#plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (7, 4)
plt.rcParams["axes.unicode_minus"] = False
```

```python
sb.barplot(data=data,
           x=data['이름'],
           y="국어",
           palette="pastel")
plt.show()
```

평균막대그래프
> 디폴트가 mean
> 막대기는 신뢰구간(95%)

```python
sb.barplot(data=data,
           x=data['학년'],
           y="국어",
           palette="pastel")
plt.show()
```

```python
sb.barplot(data=data,
           x=data['학년'],
           y="국어",
           palette="pastel",
           errorbar=("ci",98),
           estimator="max")
plt.show()
```

범주에 따른 구분 : hue

```python
sb.barplot(
    data=data,
    x="학년",
    y=data["국어"],
    hue="이름"
)
plt.legend(bbox_to_anchor=(0,1.2)) #legend 옮기기
plt.show()
```

legend 컨트롤을 위한 데이터 재배치

```python
data.set_index("이름",inplace=True)
```

```python
temps=data.columns[2:]
```

```python
type(temps[0])
```

```python
data1=data.filter(["국어"]).rename(columns={"국어":"Score"})
data1["과목"]="국어"
```

```python
data1
```

```python
result=[]
for x in temps:
    data1=data.filter([x]).rename(columns={x:"Score"})
    data1["과목"]=x
    result.append(data1)
```

```python
sb.barplot(
data=pd.concat(result),
x=pd.concat(result).index,
y="Score",
hue="과목"
)
plt.show()
```
