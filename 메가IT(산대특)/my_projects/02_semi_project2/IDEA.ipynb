```python
url='https://apis-navi.kakaomobility.com/v1/directions'
```

```python
hotel={}
error=[]
for i in range(len(data1)):
    data=session.get(url,params={'destination':f'{data1.loc[i]["경도"]},{data1.loc[i]["위도"]},name={data1.loc[i]["제목"]}',
                                     'origin':'126.4931329,33.507781,name=제주공항'},
                                    headers={'Authorization':f'KakaoAK {KEY}',
                                            'Content-Type':"json"})
    try:
        hotel[data1.loc[i]["제목"]]=data.json()["routes"][0]["summary"]
        print("okay")
    except:
        error.append(data1.loc[i]["제목"])
        print("fail")
        continue
```

```python
Big_data=[]
for ky in hotel:
    DF={}
    for i in ['destination',"fare","distance",'duration']:
        try:
            for j in hotel[ky][i].keys():
                DF[j]=hotel[ky][i][j]
        except:
            DF[i]=hotel[ky][i]
            continue
    Big_data.append(DF)
```
