```python
with open("일단저장2.txt","r",encoding="utf-8") as f:
    A=f.readlines()
```

```python
step1=A[0].split(", ")
```

```python
step2={}
for i in range(len(step1)//5):
    try:
        step2[i]=step1[5*i:5*(i+1)]
    except:
        pass
```

```python
step2
```

```python
step3={}
for i in step2:
    step3[step2[i][0].split("'")[1]]={step2[1][0].split("'")[-4]:step2[i][0].split("'")[-2]}
    step3[step2[i][0].split("'")[1]][step2[i][1].split("'")[1]]=step2[i][1].split("'")[-2]
    step3[step2[i][0].split("'")[1]][step2[i][2].split("'")[1]]=step2[i][2].split("'")[-2]
    step3[step2[i][0].split("'")[1]][step2[i][3].split("'")[1]]=step2[i][3].split("'")[-2]
    step3[step2[i][0].split("'")[1]][step2[i][4].split("'")[1]]=step2[i][4].split("'")[-2]
```

```python
step3.values()
```

```python
import pandas as pd
```

```python
pd.DataFrame(data=step3.values(),index=step3.keys()).reset_index().to_excel("일단복구완료2.xlsx")
```
