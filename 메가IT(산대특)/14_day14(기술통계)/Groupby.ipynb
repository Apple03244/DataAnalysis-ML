```python
import pandas as pd
import numpy as np
```

```python
D=pd.read_excel("https://data.hossam.kr/C02/traffic_acc.xlsx")
```

```python
data=pd.DataFrame(D)
```

```python
data.groupby(["년도","월"]).agg(["mean","sum"])
```

```python
def myfuntion(x):
    return x.max()-x.min()
```

```python
Data=data.filter(data.columns.difference(["월"])).groupby(["년도"]).agg([myfuntion])
```

#### 인덱스 해제

```python
Data.reset_index()
```

#### 인덱스를 애초에 안줄 수도 있음

```python
Data=data.filter(data.columns.difference(["월"])).groupby(["년도"],as_index=False).agg([myfuntion])
```
