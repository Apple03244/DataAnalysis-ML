```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
D=pd.read_excel("https://data.hossam.kr/E06/kings.xlsx")
data=pd.DataFrame(D)
```

```python
plt.rcParams["font.family"]="BM YEONSUNG"
plt.rcParams["figure.figsize"]=(6,5)
```

```python
data.describe().T
```

단순이동평균(SMA,Simple Moving Average)
----------

```python
#1,2,4년 기준
for year in [2,4,8]:
    data[f"sma_{year}"]=data["age"].rolling(year).mean()
```

```python
plt.rcParams["figure.figsize"]=(10,6)
_,ax=plt.subplots(2,2)
ax_l=sum(map(list,ax),[])
for i in range(3):
    sb.lineplot(data=data, x=data.index, y=data[f"sma_{2**(i+1)}"],ax=ax_l[i+1])
sb.lineplot(data=data,x=data.index,y=data["age"],ax=ax_l[0])
plt.show()
```
