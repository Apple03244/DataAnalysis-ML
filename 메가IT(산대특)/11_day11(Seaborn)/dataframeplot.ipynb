```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
D=pd.read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx")
data=pd.DataFrame(D)
data.set_index(keys="년도",drop=True,inplace=True)
```

```python
data.head(3)
```

전역설정

```python
plt.rcParams["font.family"]="BM YEONSUNG"
plt.rcParams["font.size"]=10
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['axes.unicode_minus']=False
```

#### Line Plot

```python
data["발생건수"].plot(label="교통사고")
plt.grid()
plt.legend()
plt.title("Line Plot")
plt.show()
plt.close()
```

#### 2개 이상의 시각화

```python
data.filter(["사망자수","부상자수"]).plot()
plt.grid()
plt.legend()
plt.title("Mulit Plot")
plt.show()
plt.close()
```

#### 전체 컬럼

```python
data.plot()
plt.grid()
plt.legend()
plt.title("All Plot")
plt.show()
plt.close()
```

#### Bar chart

```python
data["발생건수"].plot.bar(rot=45)
plt.grid()
plt.legend()
plt.title("Bar chart")
plt.ylim(200000,240000)
plt.show()
plt.close()
```

#### 2개 이상의 Bar Chart

```python
data.filter(data.columns.difference(["발생건수"])).plot.bar(rot=90)
plt.grid()
plt.legend()
plt.title("Multi Bar chart")
# plt.ylim(200000,240000)
plt.show()
plt.close()
```

#### 전체 Bar chart

```python
data.plot.bar(rot=90)
plt.grid()
plt.legend()
plt.title("All Bar chart")
# plt.ylim(200000,240000)
plt.show()
plt.close()
```

#### 산점도

```python
data.plot.scatter("발생건수","부상자수")
plt.grid()
# plt.legend()
plt.title("Scatter")
# plt.ylim(200000,240000)
plt.show()
plt.close()
```

```python
# 추가설명 더하기
import numpy as np
L=np.polyfit(data["발생건수"],data["부상자수"],1)
L_result=np.poly1d(L)
data["predict"]=L_result(data["발생건수"])
data
```

```python
data.plot.scatter("발생건수","부상자수")
data['predict'].plot()
plt.grid()
plt.legend()
plt.title("Bar chart")
# plt.ylim(200000,240000)
plt.show()
plt.close()
```

#### Pie chart

```python
data["사망자수"].plot.pie(autopct="%d%%")
plt.show()
plt.close()
```
