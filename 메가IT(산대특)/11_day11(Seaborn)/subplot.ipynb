```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
D=pd.read_excel("http://data.hossam.kr/D01/traffic_acc_year.xlsx")
```

```python
data=pd.DataFrame(D)
```

```python
plt.rcParams['font.family'] = 'BM YEONSUNG'
#plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (12, 8)
plt.rcParams['axes.unicode_minus'] = False
```

fig : 그래픽 처리 기능 제공
ax : 분할된 분할된 각 그래프 영역

```python
fig,ax=plt.subplots(2,2)
fig.suptitle("Subplot") #fig : super가 붙는 것
fig.subplots_adjust(wspace=0.2,hspace=0.2) #가로 세로 지정
# plt.show()
# plt.close()
data.plot(ax=ax[0][0])
data.plot.bar(ax=ax[0][1])
data["발생건수"].plot.pie(ax=ax[1][0])
data.plot.scatter(x="발생건수",y="부상자수",ax=ax[1][1])
#--------------------------------------------------------
#서브플롯 옵션 변경 예

#제목
ax[0][0].title.set_text("교통하고 건수 변화")
ax[0][0].title.set_fontsize(23)
ax[0][0].title.set_color("Red")

#x,y라벨
ax[0][1].set(xlabel="년도",ylabel="교통사고")

#xticks #인덱스 처리후 덮어씌우기가 정석인듯
ax[0][1].set_xticks(list(range(min(data.index),max(data.index)+1)))
ax[0][1].set_xticklabels(data.index,fontsize=12,color="Blue")

#x,y축 범위
ax[0][1].set_xlim(min(data.index)-0.5,max(data.index)+0.5)
'''ax[0][1].set_ylim'''

#그리드
ax[0][0].grid()

#라벨 on,off
ax[1][0].set(ylabel=None)

#legend
ax[1][0].legend(labels=data.index, title="Legend", bbox_to_anchor=(-0,1))

plt.show()
plt.close()
```

서브플롯 영역에 옵션 지정하기

>위에 정리해놨음

#### y축을 한곳에 모아 표시

```python
fig, ax1=plt.subplots(1,1,figsize=(8,4))
#쌍둥이 서브플롯의 이름=twinx
ax2=ax1.twinx() 
#----------------------------------
#ax[1]
data["발생건수"].plot(ax=ax1, color="Red")
ax1.set_xlabel("년도")
ax1.set_ylabel("발생건수",color="Red")
ax1.grid()

#ax2
data["부상자수"].plot(ax=ax2,color="Green")
ax2.set_ylabel("부상자수",color="Green")

#xticks
ax1.set_xticks(list(range((min(data.index)),max(data.index)+1)))
ax1.set_xticklabels(data["년도"],color="Blue")

plt.show()
plt.close()
```
