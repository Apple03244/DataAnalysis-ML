#### 그래프
옵션주기는 matplotlib과 같음<h>
다만 pie chart는 제공하지 않음

선 그래프

```python
sbn.lineplot(x=range(-20,20,10),y=[1,3,4,5],label="test1")
sbn.lineplot(x=range(-20,20,10),y=[1,2,3,4],label="test2")
#--------------반복--------------------#
plt.title("Line")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
plt.close()
```

```python
x=["서울","부산","제주"]
y=[10,20,30]
sbn.barplot(x=x,y=y)
#--------------반복--------------------#
plt.title("Line")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
plt.close()
```

```python
x=["서울","부산","제주"]
y=[10,20,30]
sbn.barplot(x=y,y=x)
#--------------반복--------------------#
plt.title("Bar")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
plt.close()
```

산점도

```python
tmp = [ 23, 25, 26, 27, 28, 29, 30, 31, 33 ]
qty = [ 21, 23, 25, 28, 33, 35, 36, 32, 39 ]

sbn.scatterplot(x=tmp,y=qty)
#--------------반복--------------------#
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
plt.close()
```

DataFrame을 이용한 Seaborn chart

```python
df = pd.read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx")
```

```python
sbn.lineplot(data=df,x=df.index,y="사망자수")
#--------------반복--------------------#
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(df.index,df["년도"])
plt.grid()
plt.show()
plt.close()
```
